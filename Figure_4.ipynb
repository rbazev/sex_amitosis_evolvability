# Figure 4

The code in this notebook was last tested with:

* python version 3.7.12
* numpy 1.20.3
* pandas 1.3.5
* matplotlib 3.4.2

Below we show how to generate Figure 4.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
data = pd.read_csv('data/Figure_4_data.csv')
data
```

```python
fig = plt.figure(figsize=(20, 16), dpi =200)
ax = fig.add_subplot(111, projection='3d')

f_range = np.array([-5.39, -5, -4, -3, -2, -1, -0.026])
colmap = cm.ScalarMappable(cmap =  plt.cm.RdYlGn)
colmap.set_array(f_range)

x = data['U'].values
y = data['s'].values
z = data['K'].values
f = -np.log10(data['SSD'].values)

ax.scatter(x, y, z, c=f, cmap=plt.cm.RdYlGn, marker='o', s=130)
ax.scatter(x[:30], y[:30], z[:30], color='k', marker='x', s=150)

cb = fig.colorbar(colmap, label='-log(SSD)', ax=plt.gca())

cb.ax.tick_params(axis='both', labelsize=15)
font = {'size': 20}
cb.set_label('-log(SSD)',fontdict=font)

ax.set_xlabel('Genomic mutation rate (U)', labelpad=15, fontsize=25)
ax.set_ylabel('Selection coefficient (s)', labelpad=15, fontsize=25)
ax.set_zlabel("Initially fixed loci (K)", labelpad=13, fontsize=25)

ax.set_zticks([0, 5, 10, 15, 20, 25, 30])
ax.set_xticks(np.arange(0.01, 0.11, 0.01))
ax.set_yticks(np.arange(0.01, 0.11, 0.01))

plt.setp(ax.get_xticklabels(), fontsize=16)
plt.setp(ax.get_yticklabels(), fontsize=16)
plt.setp(ax.get_zticklabels(), fontsize=16);
```
